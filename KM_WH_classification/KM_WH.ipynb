```python
import numpy as np
import re
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression

from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt

! pip install nltk
import nltk.data
nltk.download('punkt')
```

```python
X_train = wh_lyrics[20:] + km_lyrics[:-20]
X_test = wh_lyrics[:20] + km_lyrics[-20:]
y_train = pd.Series(440*[0] + 462*[1])
y_test = pd.Series(20*[0] + 20*[1])
```

```python
vectorizer = CountVectorizer(lowercase=True, stop_words='english', token_pattern='[A-Za-z]+', ngram_range=(1,1))
X_train_cv = vectorizer.fit_transform(X_train)
df_bow_sklearn_train = pd.DataFrame(X_train_cv.toarray(), columns=vectorizer.get_feature_names_out())

X_test_cv = vectorizer.transform(X_test)
df_bow_sklearn_test = pd.DataFrame(X_test_cv.toarray(), columns = vectorizer.get_feature_names_out())
df_bow_sklearn_test.head()
```

```python
log_reg = LogisticRegression()
log_reg.fit(df_bow_sklearn_train, y_train)
log_reg.score(df_bow_sklearn_test, y_test), log_reg.score(df_bow_sklearn_test, y_test)
```

```python
tfidf = TfidfVectorizer(lowercase=True, stop_words='english', token_pattern='[A-Za-z]+', ngram_range=(1,1))
X_tf_train = tfidf.fit_transform(X_train)
df_tf_train = pd.DataFrame(X_tf_train.toarray(), columns=tfidf.get_feature_names_out())

X_tf_test = tfidf.transform(X_test)
df_tf_test = pd.DataFrame(X_tf_test.toarray(), columns=tfidf.get_feature_names_out())
```

```python
log_reg.fit(df_tf_train, y_train)
log_reg.score(df_tf_test, y_test), log_reg.score(df_tf_train, y_train)
```

```python
skb_cv = SelectKBest(f_classif, k=10)
skb_tfidf = SelectKBest(f_classif, k=10)
```

```python
skb_cv.fit(df_bow_sklearn_train, y_train)
skb_tfidf.fit(df_tf_train, y_train)
```

```python
features_cv = sorted(zip(skb_cv.scores_, skb_cv.get_feature_names_out()), reverse=True)
features_tfidf = sorted(zip(skb_tfidf.scores_, skb_tfidf.get_feature_names_out()), reverse=True)
```

```python
plt.figure(figsize=(15,10))
plt.bar(x=list(range(10)), height=sorted(skb_cv.scores_, reverse=True)[:10])
plt.xticks(list(range(10)), [feat[1] for feat in features_cv][:10], rotation=45, fontsize=15);
plt.title('Most frequent words in artists\' songs with Counter Vectorizer');
```

```python
plt.figure(figsize=(15,10))
plt.bar(x=list(range(10)), height=sorted(skb_tfidf.scores_, reverse=True)[:10])
plt.xticks(list(range(10)), [feat[1] for feat in features_tfidf][:10], rotation=45, fontsize=15);
plt.title('Most frequent words in artists\' songs with Term Frequency and Inverse Document Frecuency (TF-IDF)')
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
y_pred_bow = log_reg.predict(df_bow_sklearn_test)
y_pred_tf = log_reg.predict(df_tf_test)
cm_vec = confusion_matrix(y_pred_bow, y_test)
cm_tf= confusion_matrix(y_pred_tf, y_test)

cmd_vec = ConfusionMatrixDisplay(cm_vec)
cmd_tf = ConfusionMatrixDisplay(cm_tf)
```

```python
cmd_vec.plot()
plt.title('Confusion Matrix for predictions with Counter Vectorizer');
```

```python
cmd_tf.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix for predictions with TF-IDF');
```
